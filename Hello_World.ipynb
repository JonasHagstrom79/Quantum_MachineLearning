```python
import qiskit
import sys

print(f"Python-miljö: {sys.executable}")
print(f"Qiskit-version: {qiskit.__qiskit_version__}")

# --- DEN ROBUSTA IMPORTEN ---
try:
    # Försök den vanliga vägen (som i boken)
    from qiskit import QuantumCircuit, execute, Aer
    print("✅ Standard-import fungerade!")
except ImportError:
    # Om klistret fortfarande strular, gå direkt till källan (Terra)
    print("⚠️ Standard-import misslyckades, testar direktväg...")
    from qiskit.circuit import QuantumCircuit
    from qiskit.execute_function import execute
    from qiskit import Aer
    print("✅ Direkt-import fungerade! Du kan köra koden.")

# --- TESTKÖRNING ---
print("\nBygger testkrets...")
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0,1], [0,1])

print("Startar simulatorn...")
# Vi använder Aer direkt om det går
try:
    sim = Aer.get_backend('aer_simulator')
except:
    # Fallback om Aer-objektet bråkar
    import qiskit.providers.aer as aer_provider
    sim = aer_provider.Aer.get_backend('aer_simulator')

job = execute(qc, sim, shots=100)
print(f"Resultat: {job.result().get_counts()}")
```
