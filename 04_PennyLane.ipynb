```python
import pennylane as qml
```

```python
qml.__version__
```

```python
dev = qml.device('default.qubit', wires = 2)
```

```python
def qc():
    qml.PauliX(wires = 0)
    qml.Hadamard(wires = 0)
    return qml.state()
```

```python
qcirc = qml.QNode(qc, dev) # Assemble the circuit & the device.
qcirc() # Run it!
```
