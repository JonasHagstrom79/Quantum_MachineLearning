```python
from qiskit.providers.aer import AerSimulator

sim = AerSimulator()
sim_GPU = AerSimulator(device = 'GPU')
```

```python
from qiskit import *
from qiskit.visualization import *

#import job_monitor from qiskit.providers.ibmq.job

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.measure(range(2), range(2))
job = execute(qc, sim, shots = 1024)
#job_monitor(job)
result = job.result()
counts = result.get_counts()
print(counts)
#plot_histogram(counts)
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.save_statevector() # If we want to access the state vector at any point
qc.measure(0,0)
qc.measure(1,1)

result = execute(qc, sim, shots = 1024).result()
sv = result.get_statevector()
print(sv)
counts = result.get_counts()
print(counts)
```

```python
sim_u = AerSimulator(method = 'unitary')
qc = QuantumCircuit(1)
qc.h(0)
qc.save_unitary()
result = execute(qc, sim_u).result()
U = result.get_unitary(decimals = 4)
print(U)
```
